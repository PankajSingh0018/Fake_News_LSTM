```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
## Let's setup the vocabulary size 
voc_size = 5000
```

### Text Preprocessing 

```python
# Importing the nltk library and stopwords
import nltk 
import re 
from nltk.corpus import stopwords
```

```python
# downloading the stopwords
nltk.download('stopwords')
```

```python
## Dataset preprocessing using Lemmmatization and removing stop words

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

corpus=[]

for i in range(0,len(messages)):
    review= re.sub('[^a-zA-Z]', " ", messages["title"][i]) 
    
    # lowercasing all the words
    review= review.lower()
    
    # splitting the words into a list
    review= review.split()
    
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]

    #review= [ps.stem(word) for word in review if word not in stopwords.words('english')]
    
    # Joining back the words to create a single sentence
    
    review= ' '.join(review)
    
    # appending the sentences to the corpus list
    
    corpus.append( review )
```

```python
corpus
```

```python
## onehot Encoding Representation 
onehot = [one_hot(words, voc_size)for words in corpus]
```

```python
onehot
```

```python
corpus[1]
```

```python
onehot[1]
```

```python
## Embedding Representation 
## making the sentencess of equal length 
sentence_length = 15

# Padding the sentences
embedded_docs= pad_sequences(onehot, padding='post', maxlen=sentence_length)
```

```python
embedded_docs
```

```python
# Creating the model 

embedded_vector_features = 40 ## Feature Representation / dimensions 
model = Sequential()

model.add(Embedding(voc_size, embedded_vector_features, input_length=sentence_length))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

print(model.summary())
```

```python
len(embedded_docs), y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

### Model Traning 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
### Finally Training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

#### Adding Dropouts

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sentence_length))
model.add(Dropout(0.3))
model.add(LSTM(2000))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

### Performance Metrics And Accuracy

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.6, 1,0) ##AUC ROC Curve to select the probability threshold
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
